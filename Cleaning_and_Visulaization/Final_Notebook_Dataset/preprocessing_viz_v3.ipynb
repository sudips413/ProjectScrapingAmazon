```python
df["Color"] = df["Color"].str.replace("Gray", " Grey")
```

```python
dfnew=df.groupby(['Brand','Color'])['Model_Name'].count().sort_values(ascending=False)
dfnew=dfnew.head(10)
data =dfnew.to_dict()
data
```

```python
brand_dict = {}
for brand, color in data.keys():
    if brand not in brand_dict:
        brand_dict[brand] = {}
    brand_dict[brand][color] = data[(brand, color)]

print(brand_dict)
```

```python
brands = list(brand_dict.keys())
colors = list(brand_dict[brands[0]].keys())

brand_color_values = np.zeros((len(brands), len(colors)))
for i, brand in enumerate(brands):
    for j, color in enumerate(colors):
        brand_color_values[i][j] = brand_dict[brand].get(color, 0)
        
bar_width = 0.2
x = np.arange(len(brands))

fig, ax = plt.subplots(figsize=(20,6))
for i in range(len(colors)):
    ax.bar(x + i * bar_width, brand_color_values[:, i], bar_width, label=colors[i])

ax.set_xlabel('Brand')
ax.set_ylabel('Count')
ax.set_title('Brand vs Color Count')
ax.set_xticks(x)
ax.set_xticklabels(brands)
ax.legend()

plt.show()
```

```python
fig = plt.figure(figsize =(10, 10))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
data=[data1,data5,data2,data3,data4]
# Creating plot
bp = ax.boxplot(data,labels=['Sony','Skullcandy','JBL','JVC','Sennheiser'])
plt.xlabel('Brand')
plt.ylabel('Ratings')
plt.title('Box plot for Beats Headphones') 
# show plot
plt.show()
```

```python
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
 
# Creating axes instance
bp = ax.boxplot(data, patch_artist = True,
                notch ='True', vert = 0)
 
colors = ['#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF']
 
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
 
# changing color and linewidth of
# whiskers
for whisker in bp['whiskers']:
    whisker.set(color ='#8B008B',
                linewidth = 1.5,
                linestyle =":")
 
# changing color and linewidth of
# caps
for cap in bp['caps']:
    cap.set(color ='#8B008B',
            linewidth = 2)
 
# changing color and linewidth of
# medians
for median in bp['medians']:
    median.set(color ='red',
               linewidth = 3)
 
# changing style of fliers
for flier in bp['fliers']:
    flier.set(marker ='D',
              color ='#e7298a',
              alpha = 0.5)
     
# x-axis labels
ax.set_yticklabels(['data_1', 'data_2',
                    'data_3', 'data_4'])
 
# Adding title
plt.title("Customized box plot")
 
# Removing top axes and right axes
# ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
     
# show plot
plt.show()
```
