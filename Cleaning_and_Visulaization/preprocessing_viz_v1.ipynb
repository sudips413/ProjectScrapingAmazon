################################# weight Preprocessing################################

```python
#convert the weight ounces to pounds
#replaces 

df["weight(ounces)"] = df["weight(ounces)"].str.replace("ounces", "").str.strip()
df["weight(ounces)"] = df["weight(ounces)"].str.replace("pounds", "").str.strip()
df["weight(ounces)"] = df["weight(ounces)"].where(df["weight(ounces)"] != "NaN", "unknown")
df["weight(ounces)"] = df["weight(ounces)"].where(df["weight(ounces)"] != "", "unknown")
df["weight(ounces)"] = pd.to_numeric(df["weight(ounces)"], errors="coerce") / 16
df["weight(ounces)"] = df["weight(ounces)"].fillna("unknown")
```

```python
# replace the weight(ounces) to weight(pounds)

df.rename(columns={'weight(ounces)':'weight(pounds)'}, inplace=True)
```
