```python
import csv
data = {}
with open('datav1.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        brand = row['Brand']
        color = row['Color']
        model = row['Model_Name']
        count = int(row['Count'])
        if brand not in data:
            data[brand] = {}
        if color not in data[brand]:
            data[brand][color] = []
        data[brand][color].append({model: count})

```
